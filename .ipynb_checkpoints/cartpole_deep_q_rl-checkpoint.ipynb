```python
#make a sample gym rendering
env = gym.make('CartPole-v0')
env.reset()
rewards = []
for i in range(20):
    env.render()
    state,reward,done,info = env.step(env.action_space.sample())
    
    if done:
        rewards = []
        env.reset()
        
    rewards.append(reward)
    
env.render(close=True)
env.reset()
print(rewards)
```

```python
#create a memory for experience replay as in DeepMind paper
#want to sample from experience replay as training only on steps that are subsequent leads to correlations that can be
    #undesirable to the training

from collections import deque

class Memory():
    def __init__(self,max_size):
        self.buffer = deque(maxlen=max_size)
    def add(self,experience):
        self.buffer.append(experience)
    def sample(self,batch_size):
        idx = np.random.choice(np.arange(len(self.buffer)),size=batch_size,replace=False)
        return[ self.buffer[ii] for ii in idx]
```

```python
#create the Q Network
class QNetwork():
    def __init__(self,learning_rate = 0.005, hidden_size = 15, action_size=2, state_size = 4, name='qnetwork'):
        with tf.variable_scope(name):
            #placeholders
            #will want to feed in all three placeholders to train the network
            #will want to feed in teh state (inputs_) and possible actions to generate self.output, which shows
                #what the network is currently saying the optimal policy for that state is
            self.inputs_ = tf.placeholder(tf.float32,[None,state_size],name='inputs')
            self.actions_ = tf.placeholder(tf.int32,[None],name='actions')
            one_hot_actions = tf.one_hot(self.actions_,action_size)
            self.targetQs_ = tf.placeholder(tf.float32,[None],name='targets')

            #the layers
            self.fc1 = tf.contrib.layers.fully_connected(self.inputs_,hidden_size)
            self.fc2 = tf.contrib.layers.fully_connected(self.fc1,hidden_size)
            self.output = tf.contrib.layers.fully_connected(self.fc2,action_size,
                                                           activation_fn=None) #current best policy

            #train on memory to find best policy
            self.Q = tf.reduce_sum(tf.multiply(self.output,one_hot_actions),axis=1)
            self.loss = tf.reduce_mean(tf.square(self.targetQs_-self.Q)) 
            self.opt = tf.train.AdamOptimizer(learning_rate).minimize(self.loss)
            
```

```python
#set the hyperparameters
#episode related
gamma = 0.99 #discount factor
episodes = 1000 #how many runs
max_steps = 200 #ie how many iterations in each run

#explore vs. exploit best policy
explore_start = 1.
explore_stop = 0.01
decay_rate = 0.0001 #exponential decay

#network related
learning_rate = 0.0001
hidden_size = 64

#memory related
memory_size = 10000
batch_size = 20
pre_train_length = 20 #how many memories to populate the memory with
```

```python
#initialize the network
tf.reset_default_graph()
mainQN = QNetwork(name='main', hidden_size=hidden_size, learning_rate=learning_rate)
```

```python
#initialize the memory, add some experiences to it
memory = Memory(batch_size)
env.reset()
state,reward,done,info = env.step(env.action_space.sample())
for i in range(pre_train_length):
    action = env.action_space.sample()
    next_state,reward,done,info = env.step(action)
    if done:
        #next state is all zeros as there isn't one
        next_state = np.zeros(state.shape)
        memory.add((state,action,reward,next_state))
        env.reset()
        state,reward,done,info = env.step(env.action_space.sample())
    else:
        memory.add((state,action,reward,next_state))
        state = next_state
```

```python
#train the network
rewards_list = []
saver = tf.train.Saver()
env.reset()
state, reward, done, _ = env.step(env.action_space.sample())
#loss = 0

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    steps = 0
    for ep in range(episodes):
        t = 0
        total_reward = 0
        while t < max_steps:
            #explore available actions or exploit Q Network
            explore_prob = explore_stop + (explore_start-explore_stop)*np.exp(-decay_rate*steps)
            steps += 1

            if np.random.rand() <= explore_prob:
                #explore
                action = env.action_space.sample()
            else:
                #exploit: get current state and shape it into an input for the QNetwork
                Qs = sess.run(mainQN.output,feed_dict={mainQN.inputs_:state.reshape((1,*state.shape))})
                action = np.argmax(Qs)

            next_state,reward,done,info = env.step(action)
            total_reward += reward
            
            if done:
                t = max_steps
                next_state = np.zeros(state.shape)
                memory.add((state,action,reward,next_state))
                #display progress
                print('Episode: {}'.format(ep),
                      'Total reward: {}'.format(total_reward),
                      'Training loss: {:.4f}'.format(loss),
                      'Explore P: {:.4f}'.format(explore_prob))
                rewards_list.append((ep,total_reward))
                env.reset()
                state, reward, done, _ = env.step(env.action_space.sample())
            else:
                t += 1
                memory.add((state,action,reward,next_state))
                state = next_state
                

            #sample from experience replay to train the network
            batch = memory.sample(batch_size)
            states = np.array([var[0] for var in batch])
            actions = np.array([var[1] for var in batch])
            rewards = np.array([var[2] for var in batch])
            next_states = np.array([var[3] for var in batch])

            #get the target values from the next_states
            targetQs = sess.run(mainQN.output, feed_dict={mainQN.inputs_: next_states})
            #plug the targetQs into the Q Learning formula
            episode_ends = (next_states == np.zeros(states[0].shape)).all(axis=1)
            targetQs[episode_ends] = (0, 0)
            targetQs = reward + gamma*(np.max(targetQs,axis=1))

            #train the network
            loss, _ = sess.run([mainQN.loss,mainQN.opt], feed_dict={
                mainQN.inputs_:states,
                mainQN.actions_:actions,
                mainQN.targetQs_:targetQs
            })
        
    saver.save(sess,'checkpoints/cartpole_deep_Q.ckpt')
```

```python
#visualize the network, I just copied and pasted this part
%matplotlib inline
import matplotlib.pyplot as plt

def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / N

eps, rews = np.array(rewards_list).T
smoothed_rews = running_mean(rews, 10)
plt.plot(eps[-len(smoothed_rews):], smoothed_rews)
plt.plot(eps, rews, color='grey', alpha=0.3)
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.show()
```

```python
#test the network
test_episodes = 100
test_max_steps = 400

with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('checkpoints'))
    
    for i in range(test_episodes):
        total_reward = 0
        env.reset()
        state, reward, done, _ = env.step(env.action_space.sample())
        for t in range(test_max_steps):
            env.render()
            #get the best action
            Qs = sess.run(mainQN.output,feed_dict={mainQN.inputs_:state.reshape((1,*state.shape))})
            action = np.argmax(Qs)
            next_state,reward,done,info = env.step(action)
            total_reward += reward
            if done:
                print('Steps {} '.format(t), 'Reward: {} '.format(total_reward))
                break
            else:
                state = next_state
```

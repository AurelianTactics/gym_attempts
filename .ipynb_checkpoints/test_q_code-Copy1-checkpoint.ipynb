```python
import gym
import numpy as np
import random

env = gym.make('CartPole-v0')

Q                = {}
learningRate     = 0.1
randomActionRate = 0.99
totalReward      = 0
episodes = 10000
observation_round = 100

test_obs = []

for i_episode in range(episodes):
    observation = env.reset()
    episodeReward = 0
    #normalized = map(lambda feature: round(feature, 0), observation)
    #action = env.action_space.sample()
    #env.step(action)
    
    for t in range(300):
        if t >= 199:
            print("Game solved at episode {}".format(i_episode))
        # Choose Action
        old_observation = observation
        test_obs.append(observation)
        action = env.action_space.sample()
        random_threshold = randomActionRate - i_episode*.0001
        #if(random.random() > randomActionRate ** i_episode):
        if(random.random() > random_threshold):
            bestReward = 0
            for possibleAction in [0, 1]:
                temp_state = []
                for i in observation:
                    temp_state.append(round(i*observation_round,0))
                temp_state.append(possibleAction)
                #possibleReward = Q.get(tuple(normalized + [possibleAction]), 0)
                #possibleReward = Q.get(make_state(normalized,possibleAction),0)
                possibleReward = Q.get(tuple(temp_state[2:]),0.)
                if(possibleReward > bestReward):
                    action = possibleAction
                    bestReward = possibleReward

        # Update Q State
        observation, reward, done, info = env.step(action)
        #print(old_observation,observation)
        #round(.5554,1)
        #state = [ round(observation, 1) for elem in observation ]
        temp_state = []
        for i in old_observation:
            temp_state.append(round(i*observation_round,0))
        temp_state.append(action)
        state = tuple(temp_state[2:])
        
        temp_state2 = []
        temp_state3 = []
        for i in observation:
            temp_state2.append(round(i*observation_round,0))
            temp_state3.append(round(i*observation_round,0))
        temp_state2.append(0)
        temp_state3.append(1)
        
        reward_0 = Q.get(tuple(temp_state2[2:]),0.)
        reward_1 = Q.get(tuple(temp_state3[2:]),0.)
        next_best_reward = max(reward_0,reward_1)
            
        #print(state)
        #break
        #for i in observation:
            
#        print(observation)
#         normalized = map(lambda feature: round(feature, 1), observation)

#         temp_state = []
#         for i in normalized:
#             print(i)
#             temp_state.append(i)
#         temp_state.append(action)
#         state = tuple(temp_state)
        
        #print(state)
        #break

        if done:
            if t < 300:
                Q[state] = Q.get(state, 0) + learningRate * -1.
            print("Episode finished after {} timesteps with {} reward".format(t+1, episodeReward),random_threshold)
            break
        
        Q[state] = Q.get(state, 0) + learningRate * ( reward + next_best_reward - Q.get(state, 0.) ) 
        #learningRate * reward #learningRate * ( reward - Q.get(state, 0) )
        #Q[state] = Q.get(state, 0) + learningRate * reward
        episodeReward += reward
    #break
    totalReward += episodeReward

print("Total Reward: {}".format(totalReward), totalReward/episodes)
```
